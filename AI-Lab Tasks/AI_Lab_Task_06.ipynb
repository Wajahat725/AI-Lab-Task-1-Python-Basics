```python
import numpy as np

def evaluate_misplaced(puzzle, goal):
    count = 0
    for i in range(9):
        if puzzle[i] != goal[i] and puzzle[i] != 0:
            count += 1
    return count

def print_grid(puzzle):
    for i in range(0, 9, 3):
        print(puzzle[i], puzzle[i+1], puzzle[i+2])
    print()

def solve(puzzle, goal):
    steps = [
        ('up', -3),
        ('down', 3),
        ('left', -1),
        ('right', 1)
    ]

    # Blocked positions for each move
    blocked = {
        'up': [0, 1, 2],
        'down': [6, 7, 8],
        'left': [0, 3, 6],
        'right': [2, 5, 8]
    }

    visited = []
    priority = []
    parent_map = {}  # To store parent relationships

    h = evaluate_misplaced(puzzle, goal)
    priority.append((puzzle, 0, h))
    parent_map[tuple(puzzle)] = None  # Root has no parent

    visited.append(puzzle)
    nodes_visited = 1
    iteration = 0

    print("="*50)
    print("SOLVING 8-PUZZLE PROBLEM")
    print("="*50)
    print("\nInitial Puzzle:")
    print_grid(puzzle)
    print("Goal Puzzle:")
    print_grid(goal)
    print("Starting search...\n")

    while priority:
        priority = sorted(priority, key=lambda x: x[1] + x[2])
        puzzle, g, h = priority.pop(0)

        iteration += 1
        print(f"Iteration {iteration}:")
        print_grid(puzzle)

        if puzzle == goal:
            print("\n" + "="*50)
            print("GOAL REACHED!")
            print("="*50)

            # Reconstruct the solution path
            solution_path = []
            current = tuple(puzzle)
            while current is not None:
                solution_path.append(list(current))
                current = parent_map[current]  # Get parent directly

            solution_path.reverse()

            print("\nComplete Solution Path:")
            for i, state in enumerate(solution_path):
                print(f"Step {i}:")
                print_grid(state)

            print(f"Steps to reach goal: {len(solution_path) - 1}")
            print(f"Total nodes visited: {nodes_visited}")
            print(f"Total iterations: {iteration}")

            print("\nFinal Puzzle State:")
            print_grid(puzzle)
            return

        blank = puzzle.index(0)

        for step in steps:
            move, pos_change = step

            # Check if move is valid
            if blank in blocked[move]:
                continue

            new_blank = blank + pos_change

            if 0 <= new_blank < 9:
                new_puzzle = puzzle.copy()
                new_puzzle[blank], new_puzzle[new_blank] = new_puzzle[new_blank], new_puzzle[blank]

                new_tuple = tuple(new_puzzle)
                if new_tuple not in parent_map:
                    visited.append(new_puzzle)
                    nodes_visited += 1

                    g_new = g + 1
                    h_new = evaluate_misplaced(new_puzzle, goal)

                    priority.append((new_puzzle, g_new, h_new))
                    parent_map[new_tuple] = tuple(puzzle)  # Store parent as tuple

    print("No solution found!")

# Initial State
puzzle = [2, 8, 3,
          1, 6, 4,
          7, 0, 5]

# Goal State
goal = [2, 0, 3,
        1, 8, 4,
        7, 6, 5]

solve(puzzle, goal)
```
