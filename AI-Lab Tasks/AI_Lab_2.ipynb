Bonus Point- GPA Calculator in python

```python
print("Welcome to my GPA calculator")

total_courses = int(input("Enter your total courses: "))

total_points = 0
total_credits = 0

for i in range(total_courses):
    print("\nCourse", i + 1)

    course = input("Enter your course name: ")
    credit_hours = int(input("Enter your credit hours: "))
    obtain_marks = int(input("Enter your obtained marks: "))

    if obtain_marks >= 85:
        grade_points = 4.0
    elif obtain_marks >= 80:
        grade_points = 3.7
    elif obtain_marks >= 75:
        grade_points = 3.3
    elif obtain_marks >= 70:
        grade_points = 3.0
    elif obtain_marks >= 65:
        grade_points = 2.7
    elif obtain_marks >= 60:
        grade_points = 2.3
    elif obtain_marks >= 55:
        grade_points = 2.0
    elif obtain_marks >= 50:
        grade_points = 1.7
    else:
        grade_points = 0.0

    total_points += grade_points * credit_hours
    total_credits += credit_hours

gpa = total_points / total_credits
print("\nYour GPA is:", round(gpa, 2))
```
